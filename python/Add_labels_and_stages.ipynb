# Add labels and stages of gene ids to each subtype

```python
# add labels
import pandas as pd
gene_with_NM = pd.read_csv("../data/id_label/image_values_NM_MS.csv")
NM_with_lobe = pd.read_csv("../data/id_label/lobes_gif3_with_labels.csv")
merge = gene_with_NM.merge(NM_with_lobe, left_on="original", right_on="labels")
```

```python
# EBM stage for all MS Patients
EBM_for_All = pd.read_csv("../data/id_label/EBM_Stages/EBM_for_All_MS_Patients.csv")
EBM_for_All_label = merge.merge(EBM_for_All, left_on="regions", right_on="NM_regions")
```

```python
# EBM stage for RMS Patients
EBM_for_RMS = pd.read_csv("../data/id_label/EBM_Stages/EBM_for_RMS_Patients.csv")
EBM_for_RMS_label = merge.merge(EBM_for_RMS, left_on="regions", right_on="NM_regions")
```

```python
# EBM stage for PPMS Patients
EBM_for_PPMS = pd.read_csv("../data/id_label/EBM_Stages/EBM_for_PPMS_Patients.csv")
EBM_for_PPMS_label = merge.merge(EBM_for_PPMS, left_on="regions", right_on="NM_regions")
```

# correlation analysis for All MS Patients

```python
# load the GE.feather file (This file contains all the gene expression data in the Allen Dataset.)
import feather
GE = feather.read_dataframe("../Allen_data/GE.feather")
GE.set_index('probe_id', inplace=True)
GE.head()
```

```python
#load probes for analysis
use_probes = pd.read_csv('../data/probes_for_analysis.csv')
first_n = use_probes.iloc[0:,0]
```

```python
#select only the first_n probes from GE and transpose the matrix
GE_use = GE.loc[first_n,:]
GE_use_tr = GE_use.T
GE_use_tr.set_index(GE_use_tr.index.values.astype(int), inplace=True)
```

```python
#merge with the file 'EBM_for_All_label'
merged_data_All = EBM_for_All_label.merge(GE_use_tr, left_on="well_id", right_index=True)
merged_data_All
```

```python
# run the correlation analysis
from scipy.stats import spearmanr
result = []
for probe in GE_use_tr.columns:
   r, p = spearmanr(merged_data_All.EBM_Stage, merged_data_All.loc[:,probe])
   result.append([r,p])
resultAll = pd.DataFrame(result, columns=["r","p"],
index=GE_use_tr.columns)
resultAll
```

```python
gene_name =pd.DataFrame(use_probes, columns=['probe_id','gene_symbol'])
```

```python
resultAll2 = resultAll.merge(gene_name, left_on="probe_id", right_on="probe_id")
```

```python
#clean out multi-expressed genes
by_gene = resultAll2.groupby("gene_symbol")
resultAll3 = pd.DataFrame()
for gene, frame in by_gene:
    min_p = min(frame.p)
    tmp = frame[frame.p == min_p]
    resultAll3 = resultAll3.append(tmp)
```

```python
# arrange the result from small to large
resultAll4 = resultAll3.sort_values(by=['r'], ascending = False)
resultAll4.to_csv("../data/id_label/result_for_All/resultAll4.csv")
```
